```python
# Create Dictionary to use in every report
def userlog():
    fh = open("userlog.log")
    userlogs = fh.readlines()
    
    fulldict = {}
    for line in userlogs:
        val = line.split()
        date = val[0]
        time = val[1]
        activity = val[2]
        server = val[3]
        domain = val[4]
        
        if domain in fulldict:
            if date in fulldict[domain]:
                fulldict[domain][date].append((time, activity, server))
            else:
                fulldict[domain][date] = [(time, activity, server)]
        else: 
            fulldict[domain] = {date: [(time, activity, server)]}
        sorted(fulldict[domain][date])
userlog()
```

```python
# Irresponsible Behavior Final Submission 
def i_behavior():
    irresponsible_behavior_report = open('irresponsible_behavior_report.txt', 'w')
    idict = userlog()
    total = 0
    udict = {}
    for user, date in idict.items():
        login_count = 0
        logout_count = 0
        sum = 0
        temp = {}
        for number, value in date.items():
            for values in value:
                if values[1] == 'login':
                    login_count += 1
                else:
                    logout_count += 1
            if login_count > logout_count:
                sum += 1
                total += 1
                if user in temp:
                    temp[number].append(value)
                else:
                    temp[number] = (value)
        if user in udict:
            temp[number].append(value)
        else:
            temp[number] = value
        udict[user] = [temp]
        udict[user].append(sum)
    irresponsible_behavior_report.write(f"{total} \n {udict}")
    print(total)
    print(udict)
    irresponsible_behavior_report.close()
i_behavior()
```
